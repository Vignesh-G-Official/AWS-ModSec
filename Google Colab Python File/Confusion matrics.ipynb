```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

df = pd.read_csv('./result1.csv')
df['ModSecurity Result Label'] = df['ModSecurity Result Label'].fillna(1)
df['AWS WAF Result Label'] = df['AWS WAF Result Label'].fillna(1)
df = df.dropna()
actual = df['Label']
predicted = df['ModSecurity Result Label']
TP = sum((actual == 1) & (predicted == 1))  # True Positive: actual=1, predicted=1
FP = sum((actual == 0) & (predicted == 1))  # False Positive: actual=0, predicted=1
TN = sum((actual == 0) & (predicted == 0))  # True Negative: actual=0, predicted=0
FN = sum((actual == 1) & (predicted == 0))  # False Negative: actual=1, predicted=0
cm = confusion_matrix(actual, predicted)
print(cm)
print(f"TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}")

predicted2 = df['AWS WAF Result Label']
cm2 = confusion_matrix(actual, predicted2)
print(cm2)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Allowed', 'Blocked'],
            yticklabels=['Allowed', 'Blocked'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix SQLi - ModSecurity")
plt.show()
```
